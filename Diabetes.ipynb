```python
# import necessary libraries
import pandas as pd
from sklearn.datasets import load_diabetes
import warnings
# load the data set
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
diabetes_df['target'] = diabetes.target
warnings.filterwarnings('ignore', category=pd.errors.DtypeWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)
```

```python
diabetes_df.head()
```

```python
diabetes_df.shape
```

```python
diabetes_df.fillna(diabetes_df.mean(), inplace=True)
```

```python
#check for missing values and handle them
diabetes_df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
diabetes_df.iloc[:, :-1] = imputer.fit_transform(diabetes_df.iloc[:, :-1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot
sns.pairplot(diabetes_df, hue='target')
plt.show()

# Correlation matrix
sns.heatmap(diabetes_df.corr(), annot=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_diabetes = diabetes_df.drop('target', axis=1)
y_diabetes = diabetes_df['target']
scaler = StandardScaler()
X_diabetes = scaler.fit_transform(X_diabetes)
X_train, X_test, y_train, y_test = train_test_split(X_diabetes, y_diabetes, test_size=0.2, random_state=42)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
print("Logistic Regression Accuracy on diabetes dataset: ", clf.score(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.neural_network import MLPClassifier, MLPRegressor
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
print("Decision Tree Accuracy on diabetes dataset: ", dt.score(X_test, y_test))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
print("Random Forest Accuracy on diabetes dataset: ", rf.score(X_test, y_test))
```

```python
# Support Vector Machine (SVM)
from sklearn.svm import SVC

svm = SVC()
svm.fit(X_train, y_train)
print("SVM Accuracy on diabetes dataset: ", svm.score(X_test, y_test))
```

```python
# K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
print("KNN Accuracy on diabetes dataset: ", knn.score(X_test, y_test))
```

```python
# Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
print("Gradient Boosting Accuracy on diabetes dataset: ", gb.score(X_test, y_test))
```

```python
# ANN
from sklearn.neural_network import MLPClassifier

ann = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500)
ann.fit(X_train, y_train)
print("ANN Accuracy on diabetes dataset: ", ann.score(X_test, y_test))
```
